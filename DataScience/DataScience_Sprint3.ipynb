```python
# Filtrar os dados onde Diabetes_012 é igual a 2 (casos de diabetes)
diabetes_cases = data[data['Diabetes_012'] == 2]

# Contagem de casos de diabetes por faixa etária
diabetes_by_age = diabetes_cases.groupby('Age')['Diabetes_012'].count()

# Plotar o histograma
plt.figure(figsize=(10,6))
plt.bar(diabetes_by_age.index, diabetes_by_age.values, color='skyblue')
plt.title('Distribuição dos Casos de Diabetes (Diabetes_012 = 2) por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Número de Casos de Diabetes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Matriz de correlação entre Diabetes_012 e variáveis mais significativas
correlation_matrix = data[['Diabetes_012', 'BMI', 'HighBP', 'DiffWalk', 'PhysActivity', 'Income', 'Education', 'Fruits', 'Veggies', 'GenHlth']].corr()

# Plotar o heatmap da matriz de correlação
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlação entre Diabetes_012 e variaveis significativas')
plt.show()
```

```python
# Realizando o teste ANOVA para comparar o IMC (BMI) entre os grupos de diabetes
# Dividimos o IMC (BMI) em três grupos com base nas categorias de diabetes
f_stat, p_valor = stats.f_oneway(
    data[data['Diabetes_012'] == 0]['BMI'],  # sem diabetes
    data[data['Diabetes_012'] == 1]['BMI'],  # pré-diabetes
    data[data['Diabetes_012'] == 2]['BMI']   # com diabetes
)

# Exibe o valor da estatística F e o valor-p (p-value)
print(f"Estatística F: {f_stat}, p-valor: {p_valor}")

# Nível de significância padrão (alpha) é 0.05
# Verificamos se o valor-p é menor que 0.05 para decidir se rejeitamos ou não a hipótese nula
if p_valor < 0.05:
    print("Rejeitamos a hipótese nula. Existem diferenças significativas no BMI entre os grupos de diabetes.")
else:
    print("Não rejeitamos a hipótese nula. Não há diferenças significativas no BMI entre os grupos de diabetes.")
```

```python
#Teste de hipotese usando qui-quadrado para diabetes e idade
# Cria a tabela de contingência entre as variáveis 'Diabetes_012' e 'Age'
cont_tab = pd.crosstab(data['Diabetes_012'], data['Age'])

# Aplica o teste qui-quadrado com base na tabela de contingência
# O teste calcula a estatística qui-quadrado, o p-valor, os graus de liberdade (dof), e os valores esperados
chi2, p_valor, dof, expected = stats.chi2_contingency(cont_tab)

# Exibe os resultados do teste qui-quadrado
print(f"Qui-quadrado: {chi2}, p-valor: {p_valor}, Graus de liberdade: {dof}")

# Interpretação dos resultados com base no p-valor
# Se o p-valor for menor que 0.05, rejeitamos a hipótese nula e concluímos que há associação
if p_valor < 0.05:
    print("Rejeitamos a hipótese nula. Existe associação entre Faixa Etária e Diabetes.")
else:
    print("Não rejeitamos a hipótese nula. Não há associação entre Faixa Etária e Diabetes.")
```

```python
#Teste de hipotese usando qui-quadrado para diabetes e colesterol alto
# Cria a tabela de contingência entre as variáveis 'Diabetes_012' e 'HighChol'
cont_tab = pd.crosstab(data['Diabetes_012'], data['HighChol'])

# Aplica o teste qui-quadrado com base na tabela de contingência
# O teste retorna a estatística qui-quadrado (chi2), o p-valor, os graus de liberdade (dof), e os valores esperados
chi2, p_valor, dof, expected = stats.chi2_contingency(cont_tab)

# Exibe os resultados do teste qui-quadrado
print(f"Qui-quadrado: {chi2}, p-valor: {p_valor}, Graus de liberdade: {dof}")

# Interpretação dos resultados com base no p-valor
# Se o p-valor for menor que 0.05, rejeitamos a hipótese nula e concluímos que há associação
if p_valor < 0.05:
    print("Rejeitamos a hipótese nula. Existe associação entre Diabetes e Colesterol Alto.")
else:
    print("Não rejeitamos a hipótese nula. Não há associação entre Diabetes e Colesterol Alto.")
```
