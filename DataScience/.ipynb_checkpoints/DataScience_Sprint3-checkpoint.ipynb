```python
# Realização das importação necessarias para auxiliar nas analises
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Carregando o arquivo csv em uma variavel
data = pd.read_csv('Sprint3 Diabetes.csv')

# Print dessa variavel
data
```

```python
#Print para a verificação do tipo de cada coluna
print(data.dtypes)
```

```python
# Verificação relacionada a existencia de valores nulos
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Exibindo o resumo estatístico do dataset
data_description = data.describe()

print(data_description)
```

```python
# Exibindo o resumo estatístico das colunas abreviadas no comando anterior
data1_description = data[['Veggies', 'HvyAlcoholConsump']].describe()

print(data1_description)
```

```python
# Filtrar os dados onde Diabetes_012 é igual a 2 (casos de diabetes)
diabetes_cases = data[data['Diabetes_012'] == 2]

# Contagem de casos de diabetes por faixa etária
diabetes_by_age = diabetes_cases.groupby('Age')['Diabetes_012'].count()

# Plotar o histograma
plt.figure(figsize=(10,6))
plt.bar(diabetes_by_age.index, diabetes_by_age.values, color='skyblue')
plt.title('Distribuição dos Casos de Diabetes (Diabetes_012 = 2) por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Número de Casos de Diabetes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Verificação sobre a distribuição dos casos, se são ou não diabeticos
estado = data['Diabetes_012'].value_counts().plot(kind='bar')
plt.title('Distribuição de Diabetes')
plt.show()
```

```python
# Verificação do coeficiente de variação para o BMI (IMC)
cv = (np.std(data['BMI']) / np.mean(data['BMI'])) * 100
print(f'Coeficiente de variação do BMI: {cv}%')
```

```python
# Veriifcação da distribuição dos BMI (IMC)
data['BMI'].hist(bins=20)
plt.title('Distribuição do BMI')
plt.show()
```

```python
# Criando o boxplot
plt.boxplot(data['BMI'])
plt.title('Boxplot do BMI')
plt.ylabel('BMI')
plt.show()
```

```python
# Matriz de correlação entre todas as variáveis
correlation_matrix = data.corr()

# Plotar o heatmap da matriz de correlação
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlação da base Diabetes')
plt.show()
```

```python
# Matriz de correlação entre BMI e todas as variáveis, incluindo Diabetes_012
correlation_matrix = data[['BMI', 'Diabetes_012', 'Age']].corr()

# Plotar o heatmap da matriz de correlação
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlação entre BMI e Diabetes_012')
plt.show()
```
